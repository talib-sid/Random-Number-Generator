```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
N = 1000
A = 106
C = 1283
M = 6075
```

```python
def gen_random_func(N):
    I = np.zeros(N,dtype=np.uint32)
    I[0] = 1
    for i in range (1,N):
        I[i] = (A * I[i-1] + C) % M
    return I
        
I = gen_random_func(N)
```

```python
print(I)
plt.scatter(range(N),I,c=np.random.rand(N), alpha=1)
plt.xlabel('Index')
plt.ylabel('I')
plt.show()
```

```python
plt.scatter(I[1:],I[:-1],c=np.random.rand(N-1),alpha=0.75)
# color = np.concatenate((np.repeat('blue', N-1), np.repeat('red', N)))
# plt.scatter(I[1:], I[:-1], c=color, alpha=0.75)
plt.xlabel("I_j+1")
plt.ylabel("I_j")
plt.show()
```

**Deduction**

We can see in I_i vs I_i+1, The Value I_i > I_i+1 for most cases (5/6)
and that every 5-6 consecutive values form a cluster of increasing varyinh in the range [0,600]

```python
nlist = [1, 10, 50] + [n for n in range(100, 2001, 100)]
Exps = []

for n in nlist:
    I = gen_random_func(n)
    Exp = np.sum(I) / n  
    print(Exp)
    Exps.append(Exp)

plt.plot(nlist,Exps)
plt.xlabel("N")
plt.ylabel("Expectation")
plt.title("Exp as a function of X")
plt.show()
```
